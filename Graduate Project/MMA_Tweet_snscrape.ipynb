```python
import snscrape
import snscrape.modules.twitter as sntwitter
import pandas as pd
# show all
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# User search
# Creating list to append tweet data to
text_list_Conor = []

# collect the tweets from Conor McGregor and khabib nurmagomedov because their are the main even fighters.
# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('from:TheNotoriousMMA since:2018-08-23, until:2018-10-07, lang:en').get_items()):
    if i > 100:
        break
    text_list_Conor.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.content])
    
# Creating a dataframe to load the list
tweets_Conor = pd.DataFrame(text_list_Conor, columns=["User", "Date Created", "Number of Likes", "Tweet"])
```

```python
# User search
# Creating list to append tweet data to
text_list_Conor = []

# collect the tweets from Conor McGregor and khabib nurmagomedov because their are the main even fighters.
# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('from:TheNotoriousMMA OR from:TeamKhabib since:2018-08-23, until:2018-10-07, , lang:en').get_items()):
    if i > 100:
        break
    text_list_Conor.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.content])
    
# Creating a dataframe to load the list
tweets_Conor_Khabib = pd.DataFrame(text_list_Conor, columns=["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"])
```

```python
# the keywords will be ufc229, ufc, TheNotoriousMMA, TeamKhabib
text_keywords = []

for i, tweet in enumerate(sntwitter.TwitterSearchScraper('ufc229 OR ufc OR TheNotoriousMMA OR TeamKhabib since:2018-08-23, until:2018-10-07, lang:en').get_items()):
    text_keywords.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.content])
    
text_ufc229_df = pd.DataFrame(text_keywords, columns = ["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"])
```
